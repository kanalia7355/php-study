## 課題1: 最初のPHPプログラム

### 解答

````php
<language>php</language><?php
echo "こんにちは、PHP世界！";
?>
````

**実行結果:**
```
こんにちは、PHP世界！
```

### 解説

1. **PHPタグ**: `<?php ?>` でPHPコードを囲みます
2. **echo命令**: 画面に文字を表示する命令です
3. **セミコロン**: 文の最後には必ず `;` をつけます
4. **文字列**: `"` で囲んで表現します

**よくあるミス:**
- セミコロンを忘れる
- `"` を閉じ忘れる
- PHPタグを書き忘れる

## 課題2: 変数を使った自己紹介

### 解答

````php
<language>php</language><?php
// 変数を定義
$name = "山田太郎";
$age = 25;
$hobby = "読書";

// 自己紹介文を表示
echo "私の名前は" . $name . "です。\n";
echo "年齢は" . $age . "歳です。\n";
echo "趣味は" . $hobby . "です。\n";
?>
````

**実行結果:**
```
私の名前は山田太郎です。
年齢は25歳です。
趣味は読書です。
```

### 解説

**変数のルール:**
- 変数名は `$` から始める（例: `$name`）
- 代入演算子 `=` で値を代入
- 大文字と小文字は区別される（`$name` と `$Name` は別物）

**文字列の連結:**
- `.` 演算子で文字列を連結
`- `\n` は改行を表す特殊文字

**別解（変数埋め込み）:**
```php
echo "私の名前は{$name}です。\n";
```

## 課題3: データ型の理解

### 解答

````php
<language>php</language><?php
// 各データ型の変数を作成
$favorite_food = "寿司";          // String（文字列）
$today = 4;                        // Integer（整数）
$height = 170.5;                   // Float（浮動小数点）
$is_fun = true;                    // Boolean（真偽値）

// var_dump()で型を確認
echo "好きな食べ物: ";
var_dump($favorite_food);

echo "今日の日付: ";
var_dump($today);

echo "身長: ";
var_dump($height);

echo "PHPは楽しい: ";
var_dump($is_fun);
?>
````

**実行結果:**
```
好きな食べ物: string(6) "寿司"
今日の日付: int(4)
身長: float(170.5)
PHPは楽しい: bool(true)
```

### 解説

**PHPの主なデータ型:**

| 型 | 説明 | 例 | var_dumpの出力 |
|---|---|---|---|
| string | 文字列 | `"寿司"` | `string(6) "寿司"` |
| integer | 整数 | `4` | `int(4)` |
| float | 浮動小数点数 | `170.5` | `float(170.5)` |
| boolean | 真偽値 | `true` | `bool(true)` |
| array | 配列 | `[1, 2, 3]` | `array(3) {...}` |
| NULL | 値なし | `null` | `NULL` |

**var_dump() の使い方:**
- 変数の型と値を同時に表示できる
- デバッグ時に非常に便利
- 配列の中身も確認できる

**注意点:**
 `- 文字列は必ず `"` か `'` で囲む
- 数値には `"` をつけない
- 真偽値は `true`/`false`（大文字小文字は区別しないが、小文字が一般的）

## 課題4: コメントの書き方

### 解答

````php
<language>php</language><?php
// ユーザー名を格納する変数
$user_name = "山田太郎";

// ログイン回数をカウント
$login_count = 0;

/*
 ユーザー歓迎プログラム

 このプログラムはユーザーがログインした際に
 歓迎メッセージを表示します。
*/

echo "ようこそ！";
?>
````

**実行結果:**
```
ようこそ！
```

### 解説

**コメントの種類:**

1. **一行コメント (`//`)**
   - `//` から行末までがコメントになる
   - 短い説明に適している
   
2. **複数行コメント (`/* */`)**
   - `/*` から `*/` までがコメントになる
   - 複数行にわたる説明に適している
   - プログラムの概要や関数の説明などで使用

**コメントを書くメリット:**
- コードの意図を他の人（や未来の自分）に伝えられる
- コードの可読性が向上する
- デバッグ時に一時的にコードを無効化できる

**ベストプラクティス:**
- 「何をしているか」ではなく「なぜそうするか」を書く
- コメントが古くならないように、コード変更時にはコメントも更新

## 課題5: 総合演習 - プロフィール作成

### 解答

````php
<language>php</language><?php
// プロフィール情報を変数に代入
$name = "山田太郎";              // 文字列
$age = 25;                       // 整数
$height = 170.5;                 // 浮動小数点
$hobbies = ["読書", "旅行", "プログラミング"];  // 配列
$is_student = false;             // 真偽値

// プロフィールを表示
echo "=== プロフィール ===\n";
echo "名前: " . $name . "\n";
echo "年齢: " . $age . "歳\n";
echo "身長: " . $height . "cm\n";
echo "趣味: " . $hobbies[0] . ", " . $hobbies[1] . ", " . $hobbies[2] . "\n";
echo "学生: " . ($is_student ? "はい" : "いいえ") . "\n";
echo "===================";
?>
````

**実行結果:**
```
=== プロフィール ===
名前: 山田太郎
年齢: 25歳
身長: 170.5cm
趣味: 読書, 旅行, プログラミング
学生: いいえ
===================
```

### 解説

**データ型の使い分け:**

1. **文字列（String）**: `$name` - 名前などのテキストデータ
2. **整数（Integer）**: `$age` - 年齢などの数量データ
3. **浮動小数点（Float）**: `$height` - 身長などの小数を含むデータ
4. **配列（Array）**: `$hobbies` - 複数の関連データ
5. **真偽値（Boolean）**: `$is_student` - Yes/Noの2択データ

**配列のアクセス方法:**
- `$hobbies[0]` - 1番目の要素（インデックスは0から始まる）
- `$hobbies[1]` - 2番目の要素
- `$hobbies[2]` - 3番目の要素

**三項演算子:**
```php
$is_student ? "はい" : "いいえ"
```
- 条件 `$is_student` が `true` なら「はい」
- `false` なら「いいえ」を返す

**別解（foreachループを使用）:**
```php
echo "趣味: ";
foreach ($hobbies as $index => $hobby) {
    if ($index > 0) {
        echo ", ";
    }
    echo $hobby;
}
echo "\n";
```

この方法なら、趣味の数が変わってもコードを変更する必要がありません。
