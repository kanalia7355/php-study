## 課題1: 基本的なif文

### 解答

````php
<language>php</language><?php
$number = -5;

if ($number > 0) {
    echo "正の数です\n";
} elseif ($number < 0) {
    echo "負の数です\n";
} else {
    echo "ゼロです\n";
}
?>
````

**実行結果:**
```
負の数です
```

### 解説

**if文の基本構造:**
```php
if (条件1) {
    // 条件1がtrueのときの処理
} elseif (条件2) {
    // 条件1がfalseで、条件2がtrueのとき
} else {
    // 全ての条件がfalseのとき
}
```

**重要なポイント:**
- 条件は上から順に評価される
- 最初に `true` になったブロックだけ実行される
- `elseif` は複数書ける
- `else` は省略可能

## 課題2: 年齢による区分判定

### 解答

````php
<language>php</language><?php
$age = 17;

if ($age < 0) {
    echo "無効な年齢です\n";
} elseif ($age <= 12) {
    echo "子供\n";
} elseif ($age <= 19) {
    echo "若者（ティーンエイジャー）\n";
} elseif ($age <= 64) {
    echo "成人\n";
} else {
    echo "シニア\n";
}
?>
````

**実行結果:**
```
若者（ティーンエイジャー）
```

### 解説

**境界値のテスト:**

| 年齢 | 期待する結果 | 理由 |
|------|--------------|------|
| -1 | 無効な年齢 | 負の数 |
| 0 | 子供 | 0〜12歳の範囲 |
| 12 | 子供 | 上限値 |
| 13 | 若者 | 下限値 |
| 19 | 若者 | 上限値 |
| 20 | 成人 | 下限値 |
| 64 | 成人 | 上限値 |
| 65 | シニア | 下限値 |
| 100 | シニア | 上限なし |

**コードのポイント:**

1. **バリデーション**: 最初に不正な値をチェック
   ```php
   if ($age < 0) {
       echo "無効な年齢です\n";
   }
   ```

2. **範囲チェック**: 小さい順にチェックしていく
   ```php
   elseif ($age <= 12) {  // 0〜12
   elseif ($age <= 19) {  // 13〜19
   ```

**よくある間違い:**
```php
// 良くない例: 範囲が重複している
if ($age >= 0 && $age <= 12) {
} elseif ($age >= 13 && $age <= 19) {
}

// 良い例: 範囲が明確
if ($age <= 12) {
} elseif ($age <= 19) {
}
```

## 課題3: switch文で曜日を判定

### 解答

````php
<language>php</language><?php
$day = 3;

switch ($day) {
    case 0:
        echo "日曜日です。ゆっくり休みましょう！\n";
        break;
    case 1:
        echo "今週も頑張りましょう！\n";
        break;
    case 2:
        echo "火曜日です。調子はどうですか？\n";
        break;
    case 3:
        echo "週の真ん中です！\n";
        break;
    case 4:
        echo "もうすぐ週末です！\n";
        break;
    case 5:
        echo "週末が近いです！\n";
        break;
    case 6:
        echo "週末です。楽しみましょう！\n";
        break;
    default:
        echo "無効な曜日番号です\n";
        break;
}
?>
````

**実行結果:**
```
週の真ん中です！
```

### 解説

**switch文の構造:**
```php
switch ($value) {
    case 値1:
        // $value == 値1 のときの処理
        break;  // switchから抜ける
    case 値2:
        // $value == 値2 のときの処理
        break;
    default:
        // どのcaseにも一致しないとき
}
```

**breakの重要性:**

```php
// breakを忘れるとどうなるか？
switch ($day) {
    case 1:
        echo "月曜日\n";
        // breakがない！
    case 2:
        echo "火曜日\n";
        break;
}

// $day = 1 のときの出力:
// 月曜日
// 火曜日
```

**フォールスルー（わざとbreakを書かない）:**

```php
switch ($day) {
    case 0:
    case 6:
        echo "週末です！\n";
        break;
    default:
        echo "平日です\n";
}
```

**switch vs if:**

| 特徴 | switch | if-elseif |
|------|--------|----------|
| 適した用途 | 単一の値の比較 | 複雑な条件式 |
| 可読性 | 多方向分岐で良い | 複雑な条件で良い |
| 柔軟性 | 低い（値の比較のみ） | 高い（任意の条件） |

## 課題4: 三項演算子を使った条件式

### 解答

````php
<language>php</language><?php
// 1. パスワードの強度判定
$password = "mypassword";
$strength = strlen($password) >= 8 ? "強い" : "弱い";
echo "パスワード強度: " . $strength . "\n";

// 2. 奇数・偶数判定
$number = 7;
$result = $number % 2 == 0 ? "偶数です" : "奇数です";
echo "{$number}は" . $result . "\n";

// 3. 会員価格の計算
$price = 1000;
$is_member = true;
$member_price = $is_member ? $price * 0.9 : $price;
echo "会員価格: " . $member_price . "円\n";
?>
````

**実行結果:**
```
パスワード強度: 強い
7は奇数です
会員価格: 900円
```

### 解説

**三項演算子の構文:**
```php
条件 ? 真の場合の値 : 偽の場合の値
```

**if文との比較:**

```php
// if文で書く場合
if ($is_member) {
    $member_price = $price * 0.9;
} else {
    $member_price = $price;
}

// 三項演算子で書く場合
$member_price = $is_member ? $price * 0.9 : $price;
```

**ネストした三項演算子:**

```php
// 複数の条件
$score = 75;
$grade = $score >= 80 ? "A" : ($score >= 70 ? "B" : "C");
// B
```

**使用のガイドライン:**

| 良い使用例 | 悪い使用例 |
|-----------|-----------|
| 単純な2択 | 複雑なネスト |
| 値の代入 | 複数の処理 |
| 短い条件式 | 可読性を損なう場合 |

## 課題5: 複合条件を使ったログイン認証

### 解答

````php
<language>php</language><?php
$input_username = "yamada";
$input_password = "pass123";
$valid_username = "yamada";
$valid_password = "pass123";
$is_active = true;
$is_locked = false;

// 認証ロジック
if ($input_username !== $valid_username) {
    echo "ユーザー名が間違っています\n";
} elseif ($input_password !== $valid_password) {
    echo "パスワードが間違っています\n";
} elseif (!$is_active) {
    echo "アカウントが無効です\n";
} elseif ($is_locked) {
    echo "アカウントはロックされています\n";
} else {
    echo "ログイン成功！\n";
}
?>
````

**実行結果:**
```
ログイン成功！
```

### 解説

**認証フロー:**
```
1. ユーザー名チェック
   ↓ 正しい場合
2. パスワードチェック
   ↓ 正しい場合
3. アカウント有効チェック
   ↓ 有効な場合
4. ロック状態チェック
   ↓ ロックされていない場合
5. ログイン成功
```

**重要なポイント:**

1. **チェックの順序**: 最も早く失敗する条件から
2. **厳密比較**: `!==` を使って型も比較
3. **論理否定**: `!` を使ってチェック

**セキュリティの考慮点:**

```php
// 良くない例: 攻撃者に情報を与えすぎ
if ($username !== $valid_username) {
    echo "ユーザー名が間違っています";
}

// 良い例: 一般的なメッセージ
if ($username !== $valid_username || $password !== $valid_password) {
    echo "ユーザー名またはパスワードが間違っています";
}
```

## 課題6: ネストした条件分岐 - 成績評価システム

### 解答

````php
<language>php</language><?php
$exam_score = 85;
$attendance_rate = 92;
$all_assignments_submitted = true;

// 成績を評価
if ($exam_score >= 90) {
    $grade = "A";
} elseif ($exam_score >= 80) {
    $grade = "B";
} elseif ($exam_score >= 70) {
    $grade = "C";
} elseif ($exam_score >= 60) {
    $grade = "D";
} else {
    $grade = "F";
}

// 出席率による調整
if ($grade != "A" && $grade != "F") {  // AとFは変更しない
    if ($attendance_rate >= 90) {
        // 成績を1つ上げる
        $grade_map_up = ["B" => "A", "C" => "B", "D" => "C"];
        $grade = $grade_map_up[$grade];
    } elseif ($attendance_rate < 60) {
        // 成績を1つ下げる
        $grade_map_down = ["A" => "B", "B" => "C", "C" => "D", "D" => "F"];
        $grade = $grade_map_down[$grade];
    }
}

// 課題提出によるボーナス
$bonus = $all_assignments_submitted ? "+" : "";
$final_grade = $grade . $bonus;

// 結果を表示
echo "試験点数: " . $exam_score . "点\n";
echo "出席率: " . $attendance_rate . "%\n";
echo "課題提出: " . ($all_assignments_submitted ? "完了" : "未完了") . "\n";
echo "成績: " . $final_grade . "\n";
?>
````

**実行結果:**
```
試験点数: 85点
出席率: 92%
課題提出: 完了
成績: A+
```

### 解説

**評価ロジック:**

```
試験85点 → B
出席率92% → 成績を1つ上げる (B→A)
課題完了 → +を追加
最終成績: A+
```

**ネストの深さを抑えるコツ:**

```php
// 深いネスト（悪い例）
if ($grade != "A") {
    if ($grade != "F") {
        if ($attendance_rate >= 90) {
            // 処理
        }
    }
}

// 浅いネスト（良い例）
if ($grade != "A" && $grade != "F") {
    if ($attendance_rate >= 90) {
        // 処理
    }
}
```

**早期リターンパターン:**

    if ($grade == "A" || $grade == "F") {

## 課題7: 総合演習 - 料金計算システム

### 解答

````php
<language>php</language><?php
$distance = 12.3;
$hour = 2;
$is_member = true;
$age = 70;
$is_airport = false;
$has_large_luggage = true;

// 1. 基本料金を計算
$base_fare = 500;  // 初乗り運賃
$extra_distance = $distance - 1.5;  // 超過距離

if ($extra_distance > 0) {
    $extra_units = ceil($extra_distance);  // 切り上げ
    $distance_fare = $extra_units * 300;
} else {
    $distance_fare = 0;
}

$fare = $base_fare + $distance_fare;

// 2. 割引率を決定（最も高い率を適用）
$discount_rate = 0;

// 深夜割引（23時〜5時）
if ($hour >= 23 || $hour < 5) {
    $discount_rate = max($discount_rate, 0.20);
}

// 会員割引
if ($is_member) {
    $discount_rate = max($discount_rate, 0.10);
}

// 高齢者割引
if ($age >= 65) {
    $discount_rate = max($discount_rate, 0.15);
}

// 3. 割引を適用
$discount_amount = $fare * $discount_rate;
$fare = $fare - $discount_amount;

// 4. 追加料金を加算
$extras = 0;
if ($is_airport) {
    $extras += 1000;
}
if ($has_large_luggage) {
    $extras += 500;
}

$total_fare = $fare + $extras;

// 5. 合計料金を表示
echo "=== タクシー料金明細 ===\n";
echo "走行距離: " . $distance . "km\n";
echo "初乗り運賃: " . $base_fare . "円\n";
echo "超過距離料金: " . $distance_fare . "円\n";
echo "小計: " . ($base_fare + $distance_fare) . "円\n";
echo "割引: -" . $discount_amount . "円（" . ($discount_rate * 100) . "%）\n";
echo "追加料金: " . $extras . "円\n";
echo "-------------------\n";
echo "合計料金: " . floor($total_fare) . "円\n";
?>
````

**実行結果:**
```
=== タクシー料金明細 ===
走行距離: 12.3km
初乗り運賃: 500円
超過距離料金: 3300円
小計: 3800円
割引: -760円（20%）
追加料金: 500円
-------------------
合計料金: 3540円
```

### 解説

**計算の詳細:**

```
1. 基本料金
   初乗り運賃: 500円
   超過距離: 12.3 - 1.5 = 10.8km
   超過単位: ceil(10.8) = 11単位
   超過料金: 11 × 300 = 3,300円
   小計: 500 + 3,300 = 3,800円

2. 割引適用
   深夜割引: 20%（2時台なので適用）
   会員割引: 10%
   高齢者割引: 15%
   → 最も高い20%を適用
   割引額: 3,800 × 0.20 = 760円

3. 追加料金
   大型荷物: 500円

4. 合計
   3,800 - 760 + 500 = 3,540円
```

**max()関数を使った割引率の決定:**
```php
$discount_rate = max($discount_rate, 0.20);
```
これにより、常に最も高い割引率が適用されます。

**ceil()関数で切り上げ:**
```php
ceil(10.1) = 11
ceil(10.5) = 11
ceil(10.8) = 11
```
