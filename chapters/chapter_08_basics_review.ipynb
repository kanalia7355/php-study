```php
<?php

/**
 * スロットを回す
 * @return array 3つのシンボル
 */
function spin(): array {
    $symbols = ['🍎', '🍊', '🍋', '🍇', '💎', '7️⃣'];
    
    // ランダムに3つのシンボルを選んで返してください
    // ヒント: array_rand() または rand()
    
}

/**
 * 当たり判定
 * @param array $symbols 3つのシンボル
 * @return array ['win' => bool, 'payout' => int, 'message' => string]
 */
function checkWin(array $symbols): array {
    // ここを実装してください
    // 
    // ルール:
    // - 3つすべて同じ: 100コイン
    // - 2つ同じ: 10コイン
    // - 💎が1つ以上: 5コイン
    // - 7️⃣が3つ: ジャックポット500コイン
    // 
    return [
        'win' => false,
        'payout' => 0,
        'message' => ''
    ];
}

/**
 * スロットを表示
*/
function displaySlot(array $symbols): string {
    return "\n┌─────────┐\n│ {$symbols[0]} │ {$symbols[1]} │ {$symbols[2]} │\n└─────────┘\n";
}

// ゲーム実行
$coins = 100;

echo "🎰 スロットマシーン 🎰\n";
echo "所持コイン: {$coins}\n";

while ($coins > 0) {
    echo "\nEnterキーを押してスピン（終了するなら 'q'）: ";
    $input = trim(fgets(STDIN));
    
    if ($input === 'q') {
        break;
    }
    
    if ($coins < 10) {
        echo "コインが足りません！\n";
        break;
    }
    
    $coins -= 10;
    echo "\n10コインベット！\n";
    
    $symbols = spin();
    echo displaySlot($symbols);
    
    $result = checkWin($symbols);
    
    if ($result['win']) {
        $coins += $result['payout'];
        echo "{$result['message']} +{$result['payout']}コイン！\n";
    } else {
        echo "はずれ...\n";
    }
    
    echo "所持コイン: {$coins}\n";
}

echo "\nゲーム終了！最終所持コイン: {$coins}\n";

?>
```

## 7. パフォーマンス最適化の基礎

良いコードを書くためのヒントです。

### ループ内での文字列結合

```php
<?php

// 悪い例
$result = '';
for ($i = 0; $i < 1000; $i++) {
    $result .= $i;
}

// 良い例
$parts = [];
for ($i = 0; $i < 1000; $i++) {
    $parts[] = $i;
}
$result = implode('', $parts);

?>
```

### 無駄な計算を避ける

```php
<?php

// 悪い例（毎回countを呼ぶ）
$items = [1, 2, 3, 4, 5];
for ($i = 0; $i < count($items); $i++) {
    echo $items[$i];
}

// 良い例（事前に取得）
$count = count($items);
for ($i = 0; $i < $count; $i++) {
    echo $items[$i];
}

// より良い例（foreachを使う）
foreach ($items as $item) {
    echo $item;
}

?>
```
