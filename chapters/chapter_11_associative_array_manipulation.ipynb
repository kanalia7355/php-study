## 1. 連想配列の基本操作をおさらい

連想配列は、キーと値のペアを持つデータ構造です。

```php
<?php

// 基本的な連想配列
$user = [
    'name' => '山田太郎',
    'email' => 'yamada@example.com',
    'age' => 30,
    'city' => '東京'
];

print_r($user);

// 値の取得
echo "名前: {$user['name']}\n";
echo "年齢: {$user['age']}歳\n";

// 値の変更
$user['age'] = 31;
echo "変更後の年齢: {$user['age']}歳\n";

// 要素の追加
$user['phone'] = '090-1234-5678';
print_r($user);

// 要素の削除
unset($user['phone']);
print_r($user);

// キーの存在確認
if (isset($user['email'])) {
    echo "メールアドレスは存在します\n";
}

if (array_key_exists('address', $user)) {
    echo "住所は存在します\n";
} else {
    echo "住所は存在しません\n";
}

?>
```

## 2. 多次元連想配列

連想配列の中に連想配列を入れることで、複雑なデータ構造を表現できます。

```php
<?php

// ユーザーリスト（多次元連想配列）
$users = [
    1 => [
        'id' => 1,
        'name' => '山田太郎',
        'email' => 'yamada@example.com',
        'profile' => [
            'age' => 30,
            'city' => '東京',
            'hobbies' => ['読書', '旅行', 'プログラミング']
        ]
    ],
    2 => [
        'id' => 2,
        'name' => '鈴木花子',
        'email' => 'suzuki@example.com',
        'profile' => [
            'age' => 25,
            'city' => '大阪',
            'hobbies' => ['音楽', '映画']
        ]
    ],
    3 => [
        'id' => 3,
        'name' => '佐藤次郎',
        'email' => 'sato@example.com',
        'profile' => [
            'age' => 35,
            'city' => '名古屋',
            'hobbies' => ['スポーツ', 'ゲーム', '料理', '釣り']
        ]
    ]
];

// データへのアクセス
echo "ユーザー名: {$users[1]['name']}\n";
echo "年齢: {$users[1]['profile']['age']}歳\n";
echo "趣味: " . implode('、', $users[1]['profile']['hobbies']) . "\n";

// 全ユーザーの情報を表示
foreach ($users as $user) {
    echo "\n{$user['name']}（{$user['profile']['city']}）\n";
    echo "  趣味: " . implode('、', $user['profile']['hobbies']) . "\n";
}

?>
```

## 3. 連想配列のソート

連想配列はキーまたは値でソートできます。

```php
<?php

$products = [
    'apple' => 150,
    'orange' => 100,
    'banana' => 120,
    'grape' => 300,
    'melon' => 500
];

// 値で昇順ソート（キーは維持）
asort($products);
echo "値で昇順（asort）:\n";
foreach ($products as $name => $price) {
    echo "  {$name}: {$price}円\n";
}

// 値で降順ソート
arsort($products);
echo "\n値で降順（arsort）:\n";
foreach ($products as $name => $price) {
    echo "  {$name}: {$price}円\n";
}

// キーで昇順ソート
ksort($products);
echo "\nキーで昇順（ksort）:\n";
foreach ($products as $name => $price) {
    echo "  {$name}: {$price}円\n";
}

// キーで降順ソート
krsort($products);
echo "\nキーで降順（krsort）:\n";
foreach ($products as $name => $price) {
    echo "  {$name}: {$price}円\n";
}

?>
```

## 4. 複雑な連想配列のカスタムソート

多次元連想配列を特定のキーでソートします。

```php
<?php

$students = [
    ['id' => 3, 'name' => '佐藤', 'score' => 85, 'class' => 'A'],
    ['id' => 1, 'name' => '田中', 'score' => 92, 'class' => 'B'],
    ['id' => 5, 'name' => '山田', 'score' => 78, 'class' => 'A'],
    ['id' => 2, 'name' => '鈴木', 'score' => 88, 'class' => 'C'],
    ['id' => 4, 'name' => '伊藤', 'score' => 95, 'class' => 'B']
];

// 点数で降順ソート
$byScore = $students;
usort($byScore, function($a, $b) {
    return $b['score'] <=> $a['score'];
});

echo "点数順:\n";
foreach ($byScore as $student) {
    echo "  {$student['name']}: {$student['score']}点\n";
}

// クラスと点数でソート（クラス昇順、点数降順）
$byClassAndScore = $students;
usort($byClassAndScore, function($a, $b) {
    if ($a['class'] === $b['class']) {
        return $b['score'] <=> $a['score'];  // 同じクラスなら点数で降順
    }
    return $a['class'] <=> $b['class'];  // クラスで昇順
});

echo "\nクラス・点数順:\n";
foreach ($byClassAndScore as $student) {
    echo "  {$student['class']}組 - {$student['name']}: {$student['score']}点\n";
}

// キー維持のソート（uasort）
$namedStudents = [
    'sato' => ['name' => '佐藤', 'score' => 85],
    'tanaka' => ['name' => '田中', 'score' => 92],
    'suzuki' => ['name' => '鈴木', 'score' => 78]
];

uasort($namedStudents, function($a, $b) {
    return $b['score'] <=> $a['score'];
});

echo "\nキー維持で点数順:\n";
foreach ($namedStudents as $key => $student) {
    echo "  [{$key}] {$student['name']}: {$student['score']}点\n";
}

?>
```

## 5. 連想配列のフィルタリング・変換

条件に合う要素を抽出したり、特定のキーを抽出したりします。

```php
<?php

$products = [
    ['id' => 1, 'name' => 'ノートPC', 'price' => 98000, 'category' => 'PC', 'stock' => 5],
    ['id' => 2, 'name' => 'マウス', 'price' => 1500, 'category' => '周辺機器', 'stock' => 20],
    ['id' => 3, 'name' => 'キーボード', 'price' => 3500, 'category' => '周辺機器', 'stock' => 0],
    ['id' => 4, 'name' => 'モニター', 'price' => 25000, 'category' => '周辺機器', 'stock' => 8],
    ['id' => 5, 'name' => 'タブレット', 'price' => 45000, 'category' => 'PC', 'stock' => 3]
];

// カテゴリでフィルタリング
$peripherals = array_filter($products, function($product) {
    return $product['category'] === '周辺機器';
});

echo "周辺機器:\n";
foreach ($peripherals as $product) {
    echo "  - {$product['name']}: {$product['price']}円\n";
}

// 在庫あり商品を抽出
$inStock = array_filter($products, function($product) {
    return $product['stock'] > 0;
});

echo "\n在庫あり商品: " . count($inStock) . "件\n";

// 1万円以下の商品を抽出
$affordable = array_filter($products, function($product) {
    return $product['price'] <= 10000;
});

echo "\n1万円以下:\n";
foreach ($affordable as $product) {
    echo "  - {$product['name']}: {$product['price']}円\n";
}

// 特定のキーのみ抽出（array_column）
$names = array_column($products, 'name');
echo "\n商品名: " . implode(', ', $names) . "\n";

// 特定のキーをキーとして再構築
$productsById = array_column($products, null, 'id');
echo "\nID=3の商品: {$productsById[3]['name']}\n";

// 複数のキーを抽出
$productInfo = array_map(function($product) {
    return [
        'name' => $product['name'],
        'price' => $product['price']
    ];
}, $products);

echo "\n名前と価格のみ:\n";
foreach ($productInfo as $info) {
    echo "  {$info['name']}: {$info['price']}円\n";
}

?>
```

## 6. 連想配列のグループ化・集計

特定のキーでグループ化し、集計を行います。

```php
<?php

$sales = [
    ['date' => '2026-02-01', 'product' => 'A', 'category' => '食品', 'amount' => 1000],
    ['date' => '2026-02-01', 'product' => 'B', 'category' => '飲料', 'amount' => 500],
    ['date' => '2026-02-01', 'product' => 'C', 'category' => '食品', 'amount' => 1500],
    ['date' => '2026-02-02', 'product' => 'A', 'category' => '食品', 'amount' => 2000],
    ['date' => '2026-02-02', 'product' => 'D', 'category' => '飲料', 'amount' => 800],
    ['date' => '2026-02-02', 'product' => 'B', 'category' => '飲料', 'amount' => 1000],
    ['date' => '2026-02-03', 'product' => 'C', 'category' => '食品', 'amount' => 3000],
    ['date' => '2026-02-03', 'product' => 'E', 'category' => '雑貨', 'amount' => 2500]
];

// カテゴリでグループ化
$groupedByCategory = [];
foreach ($sales as $sale) {
    $category = $sale['category'];
    if (!isset($groupedByCategory[$category])) {
        $groupedByCategory[$category] = [];
    }
        $groupedByCategory[$category][] = $sale;
}

echo "カテゴリ別グループ化:\n";
foreach ($groupedByCategory as $category => $items) {
    $total = array_sum(array_column($items, 'amount'));
    echo "  {$category}: " . count($items) . "件, {$total}円\n";
}

// 日別の売上を集計
$dailySales = [];
foreach ($sales as $sale) {
    $date = $sale['date'];
    if (!isset($dailySales[$date])) {
        $dailySales[$date] = 0;
    }
    $dailySales[$date] += $sale['amount'];
}

echo "\n日別売上:\n";
foreach ($dailySales as $date => $amount) {
    echo "  {$date}: {$amount}円\n";
}

// カテゴリ別の売上集計
$categorySales = [];
foreach ($sales as $sale) {
    $category = $sale['category'];
    if (!isset($categorySales[$category])) {
        $categorySales[$category] = ['count' => 0, 'total' => 0];
    }
    $categorySales[$category]['count']++;
    $categorySales[$category]['total'] += $sale['amount'];
}

echo "\nカテゴリ別集計:\n";
foreach ($categorySales as $category => $data) {
    $average = $data['total'] / $data['count'];
    echo "  {$category}: {$data['count']}件, 総額{$data['total']}円, 平均{$average}円\n";
}

// 商品別の売上ランキング
$productSales = [];
foreach ($sales as $sale) {
    $product = $sale['product'];
    if (!isset($productSales[$product])) {
        $productSales[$product] = 0;
    }
    $productSales[$product] += $sale['amount'];
}

arsort($productSales);  // 降順ソート

echo "\n商品別売上ランキング:\n";
$rank = 1;
foreach ($productSales as $product => $amount) {
    echo "  {$rank}位. 商品{$product}: {$amount}円\n";
    $rank++;
}

?>
```

## 7. 連想配列のマージ・_diff

複数の連想配列を結合したり、差分を取得したりします。

```php
<?php

$user1 = [
    'name' => '山田',
    'email' => 'yamada@example.com',
    'age' => 30
];

$user2 = [
    'name' => '山田',
    'email' => 'yamada.new@example.com',
    'phone' => '090-1234-5678'
];

// array_merge: 後ろの配列が優先
$merged1 = array_merge($user1, $user2);
echo "array_merge:\n";
print_r($merged1);

// + 演算子: 前の配列が優先
$merged2 = $user1 + $user2;
echo "\n+ 演算子:\n";
print_r($merged2);

// 配列の差分（値）
$array1 = [1, 2, 3, 4, 5];
$array2 = [3, 4, 5, 6, 7];
$diff1 = array_diff($array1, $array2);  // $array1にあって$array2にないもの
echo "\narray_diff:\n";
print_r($diff1);

// 連想配列の差分（キーと値）
$settings1 = ['theme' => 'dark', 'lang' => 'ja', 'fontsize' => 14];
$settings2 = ['theme' => 'dark', 'lang' => 'en'];
$settingsDiff = array_diff_assoc($settings1, $settings2);
echo "\n設定の差分:\n";
print_r($settingsDiff);

// キーの差分
$keys1 = array_keys($settings1);
$keys2 = array_keys($settings2);
$keyDiff = array_diff($keys1, $keys2);
echo "\n追加されたキー: " . implode(', ', $keyDiff) . "\n";

// 共通キーの取得
$commonKeys = array_intersect($keys1, $keys2);
echo "共通キー: " . implode(', ', $commonKeys) . "\n";

?>
```

## 8. 実践例: 設定管理システム

連想配列を使った設定管理の実装例です。

```php
<?php

/**
 * 設定管理クラス
 */
class ConfigManager {
    private array $config = [];
    private array $defaults = [];
    
    // デフォルト設定を設定
    public function setDefaults(array $defaults): void {
        $this->defaults = $defaults;
    }
    
    // 設定をロード
    public function load(array $config): void {
        // デフォルト設定とマージ（ユーザー設定が優先）
        $this->config = array_merge($this->defaults, $config);
    }
    
    // 設定値を取得
    public function get(string $key, $default = null) {
        return $this->config[$key] ?? $default;
    }
    
    // ネストした設定値を取得（ドット記法）
    public function getNested(string $path, $default = null) {
        $keys = explode('.', $path);
        $value = $this->config;
        
        foreach ($keys as $key) {
            if (!is_array($value) || !array_key_exists($key, $value)) {
                return $default;
            }
            $value = $value[$key];
        }
        
        return $value;
    }
    
    // 設定値を設定
    public function set(string $key, $value): void {
        $this->config[$key] = $value;
    }
    
    // 全設定を取得
    public function all(): array {
        return $this->config;
    }
}

// 使用例
$config = new ConfigManager();

// デフォルト設定
$config->setDefaults([
    'app' => [
        'name' => 'My App',
        'debug' => false,
        'timezone' => 'Asia/Tokyo'
    ],
    'database' => [
        'host' => 'localhost',
        'port' => 3306,
        'charset' => 'utf8mb4'
    ],
    'security' => [
        'enable_csrf' => true,
        'password_min_length' => 8
    ]
]);

// ユーザー設定で上書き
$config->load([
    'app' => [
        'debug' => true  // これだけ上書き
    ],
    'database' => [
        'host' => 'db.example.com',
        'database' => 'production_db'
    ]
]);

// 設定値を取得
echo "アプリ名: " . $config->getNested('app.name') . "\n";
echo "デバッグモード: " . ($config->getNested('app.debug') ? 'ON' : 'OFF') . "\n";
echo "DBホスト: " . $config->getNested('database.host') . "\n";
echo "タイムゾーン: " . $config->getNested('app.timezone') . "\n";

?>
```
