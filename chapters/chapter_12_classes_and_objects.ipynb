## 🎓 オブジェクト指向とは？

オブジェクト指向プログラミング（OOP）は、**「もの」を中心に考える**プログラミング手法です。

### 手続き型 vs オブジェクト指向

```php
// 手続き型：データと処理が分離している
$userName = "山田太郎";
$userAge = 30;
function celebrateBirthday($age) {
    return $age + 1;
}
$userAge = celebrateBirthday($userAge);

// オブジェクト指向：データと処理が一体化
$user = new User("山田太郎", 30);
$user->celebrateBirthday();  // 自分で誕生日を祝う
```

### オブジェクト指向の核心思想

1. **カプセル化** - データと処理を一つにまとめる
2. **抽象化** - 複雑な詳細を隠し、本質的な機能だけを公開する
3. **責任の分離** - それぞれのオブジェクトが自分の役割を果たす

**現実世界の例え**: 
- 自動車 = クラス（設計図）
- あなたの車 = オブジェクト（実体）
- 車には「アクセルを踏む」「ハンドルを切る」などの操作（メソッド）がある
- エンジンの回転数などは車の中で管理される（カプセル化）

## 1. クラスとオブジェクトの違い

| クラス | オブジェクト |
|--------|------------|
| 設計図 | 設計図から作った実物 |
| 型 | インスタンス |
| 抽象的な概念 | 具体的なデータ |
| 「車」という概念 | 「あなたの車」「私の車」 |

**重要な用語**
- **インスタンス化**: クラスからオブジェクトを作ること
- **インスタンス**: クラスから作られたオブジェクトの別名

````php
<?php

// 最もシンプルなクラス定義
class Car {
    // プロパティ（変数）
    public string $brand;
    public string $color;
    
    // メソッド（関数）
    public function drive(): void {
        echo "{$this->brand}の車が走ります！\n";
    }
    
    public function displayInfo(): void {
        echo "車種: {$this->brand}, 色: {$this->color}\n";
    }
}

// クラスからオブジェクトを作る（インスタンス化）
$myCar = new Car();
$myCar->brand = "トヨタ";
$myCar->color = "赤";

$yourCar = new Car();
$yourCar->brand = "ホンダ";
$yourCar->color = "青";

// 各オブジェクトは独立している
$myCar->displayInfo();    // 車種: トヨタ, 色: 赤
$yourCar->displayInfo();  // 車種: ホンダ, 色: 青

$myCar->drive();    // トヨタの車が走ります！
$yourCar->drive();  // ホンダの車が走ります！

// 別のオブジェクトをいくつでも作れる
$familyCar = new Car();
$familyCar->brand = "日産";
$familyCar->color = "白";
$familyCar->displayInfo();

?>
````

## 2. コンストラクタ

コンストラクタはオブジェクト作成時に自動的に呼ばれる特別なメソッドです。
初期化処理を担当します。

````php
<?php

class User {
    // プロパティ
    public string $name;
    public int $age;
    public string $email;
    
    // コンストラクタ（__constructは特別な名前）
    public function __construct(string $name, int $age, string $email) {
        echo "新しいユーザーを作成します: {$name}\n";
        
        $this->name = $name;
        $this->age = $age;
        $this->email = $email;
    }
    
    // 自己紹介メソッド
    public function introduce(): void {
        echo "こんにちは、{$this->name}です。{$this->age}歳です。\n";
    }
    
    // メールアドレスを取得
    public function getEmail(): string {
        return $this->email;
    }
}

// コンストラクタが自動的に呼ばれる
$user1 = new User("山田太郎", 30, "yamada@example.com");
$user1->introduce();

$user2 = new User("鈴木花子", 25, "suzuki@example.com");
$user2->introduce();

// $this について
// $this は「このオブジェクト自身」を指す特別な変数です
// インスタンスごとに異なる値を持てるのは $this おかげです

?>
````

## 3. プロパティの型宣言とデフォルト値

プロパティには型を宣言でき、デフォルト値も設定できます。

````php
<?php

class Product {
    // 型宣言付きのプロパティ
    public string $name;
    public int $price;
    public ?string $description;  // nullを許容
    public bool $inStock = true;  // デフォルト値
    public array $tags = [];     // デフォルト値
    
    public function __construct(string $name, int $price, ?string $description = null) {
        $this->name = $name;
        $this->price = $price;
        $this->description = $description;
    }
    
    public function displayInfo(): void {
        echo "商品名: {$this->name}\n";
        echo "価格: \{$this->price}\n";
        echo "在庫: " . ($this->inStock ? 'あり' : 'なし') . "\n";
        
        if ($this->description !== null) {
            echo "説明: {$this->description}\n";
        }
        
        if (!empty($this->tags)) {
            echo "タグ: " . implode(', ', $this->tags) . "\n";
        }
        echo "\n";
    }
}

$product1 = new Product("ノートPC", 98000, "高性能ビジネスノート");
$product1->tags = ["PC", "ビジネス", "人気"];
$product1->displayInfo();

$product2 = new Product("マウス", 1500);  // 説明なし
$product2->inStock = false;
$product2->displayInfo();

?>
````

## 4. メソッドの詳細

メソッドはクラス内で定義される関数です。

````php
<?php

class BankAccount {
    public string $owner;
    private int $balance;  // privateはクラス外からアクセス不可
    
    public function __construct(string $owner, int $initialBalance = 0) {
        $this->owner = $owner;
        $this->balance = $initialBalance;
    }
    
    // 残高を確認
    public function getBalance(): int {
        return $this->balance;
    }
    
    // 入金
    public function deposit(int $amount): void {
        if ($amount > 0) {
            $this->balance += $amount;
            echo "\{$amount}円入金しました。残高: \{$this->balance}\n";
        } else {
            echo "入金額は正の数である必要があります。\n";
        }
    }
    
    // 出金
    public function withdraw(int $amount): bool {
        if ($amount <= 0) {
            echo "出金額は正の数である必要があります。\n";
            return false;
        }
        
        if ($this->balance >= $amount) {
            $this->balance -= $amount;
            echo "\{$amount}円出金しました。残高: \{$this->balance}\n";
            return true;
        } else {
            echo "残高不足です。出金できません。\n";
            return false;
        }
    }
    
    // 残高不足かチェック
    public function isOverdrawn(int $amount): bool {
        return $this->balance < $amount;
    }
}

$account = new BankAccount("山田太郎", 10000);
echo "{$account->owner}の口座残高: \{$account->getBalance()}\n";

$account->deposit(5000);
$account->deposit(-1000);  // エラーになる

$account->withdraw(3000);
$account->withdraw(20000);  // 残高不足

// echo $account->balance;  // エラー！privateには直接アクセスできない

?>
````

## 5. 設計思想：良いクラス設計とは

### 単一責任の原則（Single Responsibility Principle）

**「一つのクラスは一つの責任だけを持つ」**

```php
// 悪い例：ユーザークラスが責務持ちすぎ
class User {
    public function register() { /* ... */ }
    public function login() { /* ... */ }
    public function sendEmail() { /* ... */ }
    public function logActivity() { /* ... */ }
    public function generateReport() { /* ... */ }
}

// 良い例：責任を分ける
class User {
    public function register() { /* ... */ }
    public function login() { /* ... */ }
}

class EmailService {
    public function send($to, $subject, $body) { /* ... */ }
}

class ActivityLogger {
    public function log($action) { /* ... */ }
}
```

````php
<?php

/**
 * 良い設計の例：タスク管理システム
 * それぞれのクラスが明確な責任を持つ
 */

// タスクのデータを表すクラス
class Task {
    private string $title;
    private bool $completed = false;
    private ?DateTime $completedAt = null;
    
    public function __construct(string $title) {
        $this->title = $title;
    }
    
    public function getTitle(): string {
        return $this->title;
    }
    
    public function isCompleted(): bool {
        return $this->completed;
    }
    
    public function complete(): void {
        if (!$this->completed) {
            $this->completed = true;
            $this->completedAt = new DateTime();
        }
    }
    
    public function undo(): void {
        $this->completed = false;
        $this->completedAt = null;
    }
}

// タスクリストを管理するクラス
class TaskList {
    private array $tasks = [];
    
    public function add(Task $task): void {
        $this->tasks[] = $task;
    }
    
    public function remove(int $index): bool {
        if (isset($this->tasks[$index])) {
            array_splice($this->tasks, $index, 1);
            return true;
        }
        return false;
    }
    
    public function get(int $index): ?Task {
        return $this->tasks[$index] ?? null;
    }
    
    public function all(): array {
        return $this->tasks;
    }
    
    public function count(): int {
        return count($this->tasks);
    }
    
    public function completedCount(): int {
        return count(array_filter($this->tasks, fn($t) => $t->isCompleted()));
    }
}

// タスクの表示を担当するクラス
class TaskPrinter {
    public function printList(TaskList $list): void {
        echo "=== タスクリスト（全{$list->count()}件）===\n";
        foreach ($list->all() as $index => $task) {
            $status = $task->isCompleted() ? '[✓]' : '[ ]';
            echo "{$index}. {$status} {$task->getTitle()}\n";
        }
        echo "完了: {$list->completedCount()}/{$list->count()}\n\n";
    }
}

// 使用例
$tasks = new TaskList();
$tasks->add(new Task("メールを返信する"));
$tasks->add(new Task("レポートを書く"));
$tasks->add(new Task("会議の準備"));

$printer = new TaskPrinter();
$printer->printList($tasks);

// タスクを完了させる
$tasks->get(0)->complete();
$tasks->get(2)->complete();

$printer->printList($tasks);

?>
````

## 6. オブジェクトの比較

PHPには2種類の比較演算子があります。

````php
<?php

class Person {
    public function __construct(
        public string $name,
        public int $age
    ) {}
}

$person1 = new Person("田中", 30);
$person2 = new Person("田中", 30);
$person3 = $person1;  // 同じインンスを参照

// == : 値が同じか（プロパティが同じか）
var_dump($person1 == $person2);  // true（プロパティが同じ）

// === : 同一インスタンスか
var_dump($person1 === $person2);  // false（別のインスタンス）
var_dump($person1 === $person3);  // true（同じインスタンス）

?>
````

## 7. クラス定数

クラスに関連する定数を定義できます。

````php
<?php

class HTTPStatus {
    // クラス定数（大文字で命名する慣習）
    public const OK = 200;
    public const CREATED = 201;
    public const NOT_FOUND = 404;
    public const SERVER_ERROR = 500;
    
    // 定数を使ったメソッド
    public static function getMessage(int $code): string {
        return match($code) {
            self::OK => "OK",
            self::CREATED => "Created",
            self::NOT_FOUND => "Not Found",
            self::SERVER_ERROR => "Internal Server Error",
            default => "Unknown Status"
        };
    }
}

// クラス定数の使い方
echo HTTPStatus::OK . "\n";  // 200
echo HTTPStatus::getMessage(HTTPStatus::NOT_FOUND) . "\n";  // Not Found

?>
````

## 8. 実践例：ECサイトの商品管理

オブジェクト指向の考え方を活かした実践的な例です。

````php
<?php

/**
 * 商品クラス
 * 責任: 商品のデータを保持し、価格計算を行う
 */
class Product {
    private string $name;
    private int $price;
    private string $category;
    private int $stock;
    
    public function __construct(string $name, int $price, string $category, int $stock) {
        $this->name = $name;
        $this->price = $price;
        $this->category = $category;
        $this->stock = $stock;
    }
    
    // ゲッター（値の取得）
    public function getName(): string {
        return $this->name;
    }
    
    public function getPrice(): int {
        return $this->price;
    }
    
    public function getCategory(): string {
        return $this->category;
    }
    
    public function getStock(): int {
        return $this->stock;
    }
    
    // 在庫チェック
    public function isInStock(): bool {
        return $this->stock > 0;
    }
    
    // 在庫を減らす
    public function decreaseStock(int $quantity): bool {
        if ($quantity <= $this->stock) {
            $this->stock -= $quantity;
            return true;
        }
        return false;
    }
    
    // 税込価格計算
    public function getPriceWithTax(float $taxRate = 0.1): int {
        return (int)round($this->price * (1 + $taxRate));
    }
}

/**
 * カートクラス
 * 責任: 商品の追加、削除、合計計算
 */
class Cart {
    private array $items = [];  // [productId => quantity]
    private array $products = [];  // [productId => Product]
    
    public function addProduct(Product $product, int $quantity = 1): bool {
        if (!$product->isInStock()) {
            echo "{$product->getName()}は在庫切れです\n";
            return false;
        }
        
        if (!$product->decreaseStock($quantity)) {
            echo "在庫が足りません\n";
            return false;
        }
        
        $id = spl_object_id($product);
        $this->products[$id] = $product;
        $this->items[$id] = ($this->items[$id] ?? 0) + $quantity;
        
        echo "{$product->getName()}を{$quantity}個カートに追加しました\n";
        return true;
    }
    
    public function removeProduct(Product $product, int $quantity = 1): void {
        $id = spl_object_id($product);
        if (isset($this->items[$id])) {
            $this->items[$id] -= $quantity;
            if ($this->items[$id] <= 0) {
                unset($this->items[$id]);
            }
        }
    }
    
    public function getSubtotal(): int {
        $total = 0;
        foreach ($this->items as $id => $quantity) {
            $product = $this->products[$id];
            $total += $product->getPrice() * $quantity;
        }
        return $total;
    }
    
    public function getTotal(float $taxRate = 0.1): int {
        return (int)round($this->getSubtotal() * (1 + $taxRate));
    }
    
    public function display(): void {
        echo "\n=== ショッピングカート ===\n";
        foreach ($this->items as $id => $quantity) {
            $product = $this->products[$id];
            $subtotal = $product->getPrice() * $quantity;
            echo "{$product->getName()} x {$quantity} = \{$subtotal}\n";
        }
        echo "----------------------\n";
        echo "小計: \{$this->getSubtotal()}\n";
        echo "合計（税込）: \{$this->getTotal()}\n\n";
    }
}

// 使用例
$cart = new Cart();

$apple = new Product("りんご", 150, "果物", 10);
$banana = new Product("ばなな", 100, "果物", 5);
$orange = new Product("オレンジ", 120, "果物", 3);

$cart->addProduct($apple, 3);
$cart->addProduct($banana, 2);
$cart->addProduct($orange, 1);

$cart->display();

?>
````
