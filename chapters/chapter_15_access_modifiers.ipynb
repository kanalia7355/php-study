## 1. アクセス修飾子とは？

アクセス修飾子とは、**クラスのプロパティやメソッドに「外部からのアクセス制限」**をかけるためのキーワードです。

### 🔍 アクセス修飾子の種類

| 修飾子 | アクセス範囲 | 説明 |
|--------|-------------|------|
| `public` | どこからでもアクセス可 | デフォルト、最も緩い制限 |
| `protected` | クラス内と継承先クラスからのみ | 一定程度の制限が必要な場合 |
| `private` | クラス内からのみ | 最も厳しい制限、完全なカプセル化 |

### 実践的な例

```php
// アクセス修飾子なし → publicと同じ
class User {
    $name = '山田太郎';
}

// 明示的に指定（推奨）
class Product {
    public $name;          // どこからでもアクセス可能
    protected $price;       // クラス内と継承先からのみ
    private $stock;        // このクラス内からのみ
}
```

````php
<language>php</language><?php

// 基本的なアクセス修飾子の例
class User {
    // public: どこからでもアクセス可能
    public $name = '山田太郎';
    public $age = 25;
    
    // protected: このクラスと継承したクラスからのみアクセス可能
    protected $email = 'yamada@example.com';
    
    // private: このクラス内からのみアクセス可能
    private $password = 'secret123';
    
    // publicメソッド: 外部から呼び出せる
    public function introduce() {
        echo "名前: {$this->name}\n";
        echo "年齢: {$this->age}\n";
        $this->showEmail(); // privateメソッドの呼び出し
    }
    
    // protectedメソッド: このクラス内でのみ呼び出せる
    protected function setEmail($email) {
        $this->email = $email;
        echo "メールアドレスを{$email}に設定しました\n";
    }
    
    // privateメソッド: このクラス内でのみ呼び出せる
    private function showEmail() {
        echo "メール: {$this->email}\n";
    }
}

// インスタンスを作成
$user = new User();

// publicプロパティにアクセス
echo "ユーザー名: {$user->name}\n"; // OK
echo "年齢: {$user->age}\n";         // OK

// protectedプロパティにアクセス（エラーになる）
// echo $user->email; // Fatal error: Cannot access protected property

// privateプロパティにアクセス（エラーになる）
// echo $user->password; // Fatal error: Cannot access private property

// publicメソッドの呼び出し
$user->introduce();

// protectedメソッドの呼び出し（エラーになる）
// $user->setEmail('new@example.com'); // Fatal error: Call to protected method

?>
````

## 2. 継承とアクセス修飾子

継承関係でのアクセス修飾子の挙動を見てみましょう。

### 継承時のルール

- `public` → 継承先でも`public`としてアクセス可能
- `protected` → 継承先では`protected`としてアクセス可能
- `private` → 継承先からは**一切アクセス不可**

````php
<language>php</language><?php

class Animal {
    public $name = '動物';
    protected $age = 0;
    private $health = 100;
    
    public function eat() {
        echo "{$this->name}は餌を食べます\n";
    }
    
    protected function sleep() {
        echo "{$this->name}は眠ります\n";
    }
    
    private function heal() {
        echo "{$this->name}の健康状態: {$this->health}\n";
    }
}

// DogクラスがAnimalを継承
class Dog extends Animal {
    public function bark() {
        echo "{$this->name}はワンワンと鳴きます！\n";
        $this->sleep(); // 継承したクラスからprotectedメソッドを呼び出し可能
        // $this->heal(); // privateメソッドはアクセス不可
    }
    
    public function setAge($age) {
        $this->age = $age; // protectedプロパティにアクセス可能
        echo "年齢を{$age}歳に設定しました\n";
    }
}

$dog = new Dog();
$dog->name = 'ポチ'; // publicプロパティは変更可能
$dog->eat(); // 継承したメソッドの呼び出し
$dog->bark();
$dog->setAge(3);

// DogクラスからはAnimalのprivateプロパティにアクセスできない
// echo $dog->health; // エラー

?>
````

## 3. カプセル化を実装する

カプセル化とは、**データ（プロパティ）と操作（メソッド）を一つにまとめ、外部から直接操作できないようにすること**です。

### ✅ 良い例：カプセル化されたクラス

````php
<language>php</language><?php

// カプセル化の良い例
class BankAccount {
    // プロパティはすべてprivateにして外部から直接操作できないようにする
    private $balance = 0;
    private $accountNumber;
    private $owner;
    
    public function __construct($accountNumber, $owner) {
        $this->accountNumber = $accountNumber;
        $this->owner = $owner;
    }
    
    // publicメソッドを介して操作を提供
    public function deposit($amount) {
        if ($amount > 0) {
            $this->balance += $amount;
            echo "{$amount}円を預金しました\n";
        } else {
            echo "預金金額は0より大きい必要があります\n";
        }
    }
    
    public function withdraw($amount) {
        if ($amount > 0 && $this->balance >= $amount) {
            $this->balance -= $amount;
            echo "{$amount}円を引き出しました\n";
            return true;
        } else {
            echo "残高不足または不正な金額です\n";
            return false;
        }
    }
    
    // 残高を取得するだけのメソッド（getter）
    public function getBalance() {
        return $this->balance;
    }
    
    public function getAccountInfo() {
        echo "口座番号: {$this->accountNumber}\n";
        echo "名義人: {$this->owner}\n";
        echo "残高: {$this->balance}円\n";
    }
}

$account = new BankAccount('12345678', '山田太郎');
$account->deposit(10000);
$account->withdraw(3000);
$account->getAccountInfo();

// 直接残高を変更しようとするとエラーになる
// $account->balance = 1000000; // Fatal error

?>
````

## 4. ゲッター（Getter）とセッター（Setter）

privateなプロパティにアクセスするための「ゲッター」と「セッター」を実装します。

### ゲッター/セッターの役割
- **Getter**: privateなデータを取得するためのメソッド
- **Setter**: privateなデータを設定するためのメソッド

````php
<language>php</language><?php

class Product {
    private $name;
    private $price;
    private $description;
    
    public function __construct($name, $price, $description = '') {
        $this->setName($name);
        $this->setPrice($price);
        $this->description = $description;
    }
    
    // Getterメソッド
    public function getName() {
        return $this->name;
    }
    
    public function getPrice() {
        return $this->price;
    }
    
    public function getDescription() {
        return $this->description;
    }
    
    // Setterメソッド（バリデーション付き）
    public function setName($name) {
        if (empty($name)) {
            throw new InvalidArgumentException('商品名は必須です');
        }
        $this->name = $name;
    }
    
    public function setPrice($price) {
        if (!is_numeric($price) || $price < 0) {
            throw new InvalidArgumentException('価格は0以上の数値で指定してください');
        }
        $this->price = (int)$price;
    }
    
    public function setDescription($description) {
        $this->description = $description;
    }
}

// 商品を作成
$product = new Product('PHP入門書', 2980, 'PHPを基礎から学べる一冊');

// Getterで値を取得
echo "商品名: {$product->getName()}\n";
echo "価格: {$product->getPrice()}円\n";
echo "説明: {$product->getDescription()}\n";

// Setterで値を変更
$product->setPrice(2500);
$product->setDescription('大幅改訂版！');

echo "\n【更新後】\n";
echo "商品名: {$product->getName()}\n";
echo "価格: {$product->getPrice()}円\n";

// 不正な値を設定しようとすると例外が発生
try {
    $product->setPrice(-100); // InvalidArgumentException
} catch (InvalidArgumentException $e) {
    echo "エラー: {$e->getMessage()}\n";
}

?>
````

## 5. マジックメソッド

PHPには、**特別な意味を持つ「マジックメソッド」**があります。

### 主なマジックメソッド

| メソッド名 | 役割 |
|------------|------|
| `__construct()` | コンストラクタ（インスタンス作成時） |
| `__destruct()` | デストラクタ（インスタンス破棄時） |
| `__toString()` | 文字列としてキャストされたとき |
| `__get($name)` | 存在しないプロパティにアクセスしたとき |
| `__set($name, $value)` | 存在しないプロパティに値をセットしたとき |

````php
<language>php</language><?php

class Smartphone {
    private $brand;
    private $model;
    private $apps = [];
    
    public function __construct($brand, $model) {
        $this->brand = $brand;
        $this->model = $model;
        echo "{$brand} {$model}を作成しました\n";
    }
    
    public function __destruct() {
        echo "{$this->brand} {$this->model}を破棄します\n";
    }
    
    public function __toString() {
        return "スマートフォン: {$this->brand} {$this->model} (アプリ数: " . count($this->apps) . ")";
    }
    
    // 存在しないプロパティにアクセスしたときに呼ばれる
    public function __get($name) {
        echo "警告: '{$name}'プロパティは存在しません\n";
        return null;
    }
    
    // 存在しないプロパティに値をセットしたときに呼ばれる
    public function __set($name, $value) {
        if ($name === 'apps') {
            $this->apps = $value;
            echo "アプリリストを更新しました\n";
        } else {
            echo "警告: '{$name}'プロパティは存在しません\n";
        }
    }
    
    public function installApp($app) {
        $this->apps[] = $app;
        echo "{$app}をインストールしました\n";
    }
    
    public function listApps() {
        echo "インストール済みアプリ:";
        foreach ($this->apps as $app) {
            echo " {$app}";
        }
        echo "\n";
    }
}

// スマートフォンを作成
$phone = new Smartphone('iPhone', '15 Pro');

// toStringのテスト
echo $phone . "\n";

// アプリをインストール
$phone->installApp('LINE');
$phone->installApp('Slack');
$phone->listApps();

// 存在しないプロパティにアクセス（__getが呼ばれる）
echo $phone->color; // 存在しないプロパティ

// 存在しないプロパティに値をセット（__setが呼ばれる）
$phone->apps = ['Instagram', 'Twitter'];
echo $phone . "\n";

?>
````

## 6. readonlyプロパティ（PHP 8.1以降）

PHP 8.1からは、**一度設定すると変更できない`readonly`プロパティ**が使えるようになりました。

readonlyプロパティの特徴：
- 1度しか設定できない（コンストラクタ内のみ）
- 自動でprivateとなる
- 再代入するとエラーになる

````php
<language>php</language><?php

// readonlyプロパティの例
class UserAccount {
    readonly string $id;
    readonly string $email;
    public string $name;
    
    public function __construct(string $id, string $email, string $name) {
        // readonlyプロパティはコンストラクタ内で設定する
        $this->id = $id;
        $this->email = $email;
        $this->name = $name;
    }
    
    // readonlyプロパティは変更できない
    public function setEmail($newEmail) {
        // $this->email = $newEmail; // Fatal error
        echo "メールアドレスは変更できません\n";
    }
}

// 作成
$user = new UserAccount('user123', 'user@example.com', '田中花子');
echo "ID: {$user->id}\n";
echo "メール: {$user->email}\n";
echo "名前: {$user->name}\n";

// readonlyプロパティの変更を試みる
// $user->id = 'new123'; // Fatal error

// 通常プロパティは変更可能
$user->name = '佐藤花子';
echo "名前（変更後）: {$user->name}\n";

?>
````

### 問題2: ショッピングカート

カートに商品を追加・削除できる`ShoppingCart`クラスを作成してください。

- 商品は配列で管理（商品名、価格）
- 合計金額を計算するメソッド
- 商品の重複チェック機能
- 数量指定機能

<language>php</language><?php
class ShoppingCart {
    private array $items = [];
    
    public function addItem(string $name, int $price, int $quantity = 1): void {
        if ($quantity <= 0) {
            echo "数量は1以上である必要があります\n";
            return;
        }
        
        $key = $name; // 簡単なため商品名をキーに
        
        if (isset($this->items[$key])) {
            // 商品が既に存在する場合は数量を加算
            $this->items[$key]['quantity'] += $quantity;
        } else {
            // 新規商品として追加
            $this->items[$key] = [
                'name' => $name,
                'price' => $price,
                'quantity' => $quantity
            ];
        }
        
        echo "{$name}を{$quantity}個カートに追加しました\n";
    }
    
    public function removeItem(string $name): void {
        if (isset($this->items[$name])) {
            unset($this->items[$name]);
            echo "{$name}をカートから削除しました\n";
        } else {
            echo "{$name}はカートに存在しません\n";
        }
    }
    
    public function getTotal(): int {
        $total = 0;
        foreach ($this->items as $item) {
            $total += $item['price'] * $item['quantity'];
        }
        return $total;
    }
    
    public function displayCart(): void {
        echo "【カート内容】\n";
        if (empty($this->items)) {
            echo "カートは空です\n";
            return;
        }
        
        foreach ($this->items as $item) {
            echo "{$item['name']} - {$item['price']}円 x {$item['quantity']} = " . 
                 ($item['price'] * $item['quantity']) . "円\n";
        }
        
        echo "\n合計金額: {$this->getTotal()}円\n";
    }
}
?>

### 問題3: ログインシステム

ユーザー認証を行う`Auth`クラスを作成してください。

- プロパティ: `users`（private, 配列）, $currentUser
- メソッド:
  - `register($username, $password)`: ユーザー登録
  - `login($username, $password)`: ログイン
  - `logout()`: ログアウト
  - `isLoggedIn()`: ログイン状態チェック

<language>php</language><?php
class Auth {
    private array $users = [];
    private ?array $currentUser = null;
    
    public function register(string $username, string $password): bool {
        if (empty($username) || empty($password)) {
            echo "ユーザー名とパスワードは必須です\n";
            return false;
        }
        
        if (isset($this->users[$username])) {
            echo "そのユーザー名は既に使用されています\n";
            return false;
        }
        
        $this->users[$username] = [
            'username' => $username,
            'password' => password_hash($password, PASSWORD_DEFAULT)
        ];
        
        echo "{$username}を登録しました\n";
        return true;
    }
    
    public function login(string $username, string $password): bool {
        if (!isset($this->users[$username])) {
            echo "ユーザーが見つかりません\n";
            return false;
        }
        
        if (password_verify($password, $this->users[$username]['password'])) {
            $this->currentUser = $this->users[$username];
            echo "{$username}としてログインしました\n";
            return true;
        } else {
            echo "パスワードが違います\n";
            return false;
        }
    }
    
    public function logout(): void {
        if ($this->currentUser) {
            echo "{$this->currentUser['username']}をログアウトしました\n";
            $this->currentUser = null;
        } else {
            echo "ログインしていません\n";
        }
    }
    
    public function isLoggedIn(): bool {
        return $this->currentUser !== null;
    }
    
    public function getCurrentUsername(): ?string {
        return $this->currentUser ? $this->currentUser['username'] : null;
    }
}
?>

### 問題4: ユーザー権限管理

ユーザーの権限（admin, editor, viewer）を管理するシステムを作成してください。

- `User`クラスに`role`プロパティをprivateで追加
- 権限に応じてアクセス可能なメソッドを切り替える
- `isAdmin()`, `isEditor()`, `isViewer()`メソッドを実装

<language>php</language><?php
class User {
    private string $username;
    private string $role; // admin, editor, viewer
    
    public function __construct(string $username, string $role) {
        $this->username = $username;
        $this->role = $role;
    }
    
    // Getterメソッド
    public function getUsername(): string {
        return $this->username;
    }
    
    public function getRole(): string {
        return $this->role;
    }
    
    // 権限チェックメソッド
    public function isAdmin(): bool {
        return $this->role === 'admin';
    }
    
    public function isEditor(): bool {
        return $this->role === 'editor' || $this->isAdmin();
    }
    
    public function isViewer(): bool {
        return $this->role === 'viewer' || $this->isEditor();
    }
    
    // 権限に応じて可能な操作
    public function canDelete(): bool {
        return $this->isAdmin();
    }
    
    public function canEdit(): bool {
        return $this->isEditor();
    }
    
    public function canView(): bool {
        return $this->isViewer();
    }
    
    public function getPermissions(): array {
        $permissions = ['view' => $this->canView()];
        if ($this->canEdit()) {
            $permissions['edit'] = true;
        }
        if ($this->canDelete()) {
            $permissions['delete'] = true;
        }
        return $permissions;
    }
}
?>
