## 1. 関数とは？

関数とは、**「まとまった処理に名前をつけて、いつでも呼び出せるようにしたもの」**です。

### 🔍 関数を使うメリット

1. **コードの再利用** - 何度も同じ処理を書かなくてよい
2. **可読性向上** - 処理の意味が分かりやすい
3. **保守性向上** - 修正が1箇所で済む
4. **テスト容易** - 個別の機能をテストしやすい

### 実践的な例

```php
// 関数なし：同じ処理を何度も書く
$price1 = 1000;
$tax1 = $price1 * 0.1;
$total1 = $price1 + $tax1;

$price2 = 2500;
$tax2 = $price2 * 0.1;
$total2 = $price2 + $tax2;

// 関数あり：すっきり！
function calculateTotalWithTax($price) {
    $tax = $price * 0.1;
    return $price + $tax;
}

$total1 = calculateTotalWithTax(1000);
$total2 = calculateTotalWithTax(2500);
```

## 2. 組み込み関数を使ってみる

PHPには最初から使える便利な関数がたくさんあります。

### よく使う組み込み関数

````php
<?php

// 文字列の長さを取得
$text = "こんにちは";
echo strlen($text) . "\n";

// 文字列の一部を取得
echo substr("Hello World", 0, 5) . "\n";

// 配列の要素数を取得
$fruits = ["りんご", "ばなな", "オレンジ"];
echo count($fruits) . "\n";

// 現在の日時を取得
echo date("Y-m-d H:i:s") . "\n";

// 数値の丸め処理
echo round(3.14159, 2) . "\n";  // 小数点第2位で丸める
echo ceil(3.1) . "\n";          // 切り上げ
echo floor(3.9) . "\n";         // 切り捨て

?>
````

## 3. 自作関数の定義

### 基本構文

```php
function 関数名(引数1, 引数2, ...) {
    // 処理
    return 戻り値;
}
```

````php
<?php

// 引数なし・戻り値なしの関数
function sayHello() {
    echo "こんにちは！\n";
}

sayHello();

// 引数ありの関数
function greet($name) {
    echo "こんにちは、{$name}さん！\n";
}

greet("田中");
greet("鈴木");

?>
````

## 4. 戻り値を返す関数

`return`を使うことで、関数の処理結果を呼び出し元に返せます。

````php
<?php

// 円の面積を計算する関数
function calculateCircleArea($radius) {
    $area = $radius * $radius * 3.14159;
    return $area;
}

$result = calculateCircleArea(5);
echo "半径5の円の面積: {$result}\n";

// 消費税込みの金額を計算
function withTax($price, $taxRate = 0.1) {
    return $price * (1 + $taxRate);
}

echo "1000円（税込）: " . withTax(1000) . "円\n";
echo "5000円（税込8％）: " . withTax(5000, 0.08) . "円\n";

?>
````

## 5. 複数の戻り値を返す

PHPでは配列を使って複数の値を返せます。

````php
<?php

// 統計情報を計算
function calculateStats($numbers) {
    $total = array_sum($numbers);
    $count = count($numbers);
    $average = $total / $count;
    $max = max($numbers);
    $min = min($numbers);
    
    return [
        'total' => $total,
        'average' => $average,
        'max' => $max,
        'min' => $min
    ];
}

$scores = [80, 95, 70, 85, 90];
$stats = calculateStats($scores);

echo "合計: {$stats['total']}\n";
echo "平均: {$stats['average']}\n";
echo "最高: {$stats['max']}\n";
echo "最低: {$stats['min']}\n";

?>
````

## 6. 型宣言（PHP 7以降）

引数と戻り値の型を宣言できます。型安全なコードを書けます。

````php
<?php

// 型宣言付きの関数
function add(int $a, int $b): int {
    return $a + $b;
}

echo add(10, 20) . "\n";

// 文字列を結合する関数
function concatenate(string $str1, string $str2): string {
    return $str1 . $str2;
}

echo concatenate("Hello, ", "World!") . "\n";

// 配列を受け取って返す関数
function filterPositive(array $numbers): array {
    return array_filter($numbers, function($n) {
        return $n > 0;
    });
}

$nums = [-1, 2, -3, 4, -5, 6];
$positives = filterPositive($nums);
print_r($positives);

?>
````

## 7. 変数のスコープ

### スコープとは
変数が有効な範囲のことです。

- **ローカル変数**: 関数内でのみ有効
- **グローバル変数**: プログラム全体で有効
- **スーパーグローバル**: どこからでもアクセス可能

````php
<?php

$globalVar = "グローバル変数";

function testScope() {
    $localVar = "ローカル変数";
    echo "関数内: " . $localVar . "\n";
    
    // グローバル変数は直接アクセスできない
    // echo $globalVar;  // エラー！
    
    // globalキーワードでアクセス
    global $globalVar;
    echo "関数内からグローバル: " . $globalVar . "\n";
}

testScope();
echo "関数外: " . $globalVar . "\n";
// echo $localVar;  // エラー！関数外からはアクセスできない

?>
````

## 8. 実践的な関数の例

### 例1: フォーム入力のバリデーション

````php
<?php

function validateEmail(string $email): bool {
    return filter_var($email, FILTER_VALIDATE_EMAIL) !== false;
}

function validateRequired(array $data, array $fields): array {
    $errors = [];
    foreach ($fields as $field) {
        if (empty($data[$field])) {
            $errors[] = "{$field}は必須です";
        }
    }
    return $errors;
}

// 使用例
$formData = [
    'name' => '山田太郎',
    'email' => 'test@example.com',
    'message' => ''
];

$errors = validateRequired($formData, ['name', 'email', 'message']);
if (!empty($errors)) {
    echo "バリデーションエラー:\n";
    foreach ($errors as $error) {
        echo "- {$error}\n";
    }
}

if (validateEmail($formData['email'])) {
    echo "メールアドレスは有効です\n";
}

?>
````

### 例2: データフォーマット変換

````php
<?php

function formatDate(string $date, string $format = 'Y年m月d日'): string {
    $timestamp = strtotime($date);
    return date($format, $timestamp);
}

function formatPrice(int $price): string {
    return number_format($price) . '円';
}

function truncateText(string $text, int $length = 100, string $suffix = '...'): string {
    if (mb_strlen($text) <= $length) {
        return $text;
    }
    return mb_substr($text, 0, $length) . $suffix;
}

// 使用例
echo formatDate('2026-02-04') . "\n";
echo formatDate('2026-02-04', 'Y/m/d') . "\n";
echo formatPrice(1234567) . "\n";
echo truncateText('これはとても長いテキストです。途中で切り取られます。', 20) . "\n";

?>
````

## 9. 無名関数（クロージャ）

関数を変数に代入したり、引数として渡せます。

````php
<?php

// 無名関数を変数に代入
$greet = function($name) {
    return "こんにちは、{$name}さん！";
};

echo $greet("田中") . "\n";

// 配列のコールバック関数として使用
$numbers = [1, 2, 3, 4, 5];

// 各要素を2倍にする
$doubled = array_map(function($n) {
    return $n * 2;
}, $numbers);

print_r($doubled);

// 偶数だけ抽出
$evens = array_filter($numbers, function($n) {
    return $n % 2 === 0;
});

print_r($evens);

?>
````

## 10. 再帰関数

関数が自分自身を呼び出す再帰処理です。

````php
<?php

// 階乗を計算
function factorial(int $n): int {
    if ($n <= 1) {
        return 1;
    }
    return $n * factorial($n - 1);
}

echo "5の階乗: " . factorial(5) . "\n";

// ディレクトリを再帰的に探索
function listFiles(string $dir, array &$results = []): array {
    $files = scandir($dir);
    foreach ($files as $file) {
        if ($file === '.' || $file === '..') {
            continue;
        }
        $path = $dir . '/' . $file;
        if (is_dir($path)) {
            listFiles($path, $results);
        } else {
            $results[] = $path;
        }
    }
    return $results;
}

// 使用例（実際のパスに合わせて調整してください）
// $files = listFiles('.');
// print_r($files);

?>
````
