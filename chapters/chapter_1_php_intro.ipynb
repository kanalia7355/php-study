## 1. PHPとは？

PHPは「**P**HP: **H**ypertext **P**reprocessor」の略で、サーバーサイドで動作するプログラミング言語です。

### 🔍 PHPの特徴

1. **ウェブ開発向け**
   - HTMLに直接埋め込んで書ける
   - 動的なウェブページを作成できる

2. **オープンソース**
   - 無料で使える
   - コミュニティが活発

3. **クロスプラットフォーム**
   - Windows, Linux, macOSなどで動作
   - 多くのウェブサーバーでサポート

4. **簡単に始められる**
   - 構文が比較的簡単
   - 初学者にも扱いやすい

## 3. PHPの基本構文

### 最初のプログラム

````php
<language>php</language><?php
echo "こんにちは、PHP世界！";
?>
````

**実行結果:**
```
こんにちは、PHP世界！
```

### 構文のポイント

- `<?php ?>` でPHPコードを囲む
- 文の最後はセミコロン（;）で終える
- コメントは `//` または `/* */` を使う

````php
<language>php</language><?php
// 一行コメント
/*
  複数行コメント
*/
echo "コメントのテスト";
?>
````

**実行結果:**
```
コメントのテスト
```

## 4. 変数とデータ型

### 変数の宣言

````php
<language>php</language><?php
$name = "山田太郎";
$age = 25;
$height = 175.5;
$is_student = true;

// 変数の値を表示
echo "名前: " . $name . "\n";
echo "年齢: " . $age . "歳\n";
echo "身長: " . $height . "cm\n";
echo "学生: " . ($is_student ? "はい" : "いいえ") . "\n";
?>
````

**実行結果:**
```
名前: 山田太郎
年齢: 25歳
身長: 175.5cm
学生: はい
```

### 主なデータ型

| 型 | 説明 | 例 |
|----|------|-----|
| string | 文字列 | `"Hello"` |
| integer | 整数 | `42` |
| float | 浮動小数点数 | `3.14` |
| boolean | 真偽値 | `true`, `false` |
| array | 配列 | `[1, 2, 3]` |
